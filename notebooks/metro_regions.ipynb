## Purpose

This notebook builds the **city-to-metro-region lookup table** used by the data cleaning notebooks (`02_fintech_data_cleaning` and `03_AI_data_cleaning`) to assign each startup to one of four major U.S. metro areas based on its Crunchbase `Organization Location` field.

Startups located outside these four metros are assigned to an implicit "Other" category during the cleaning merge (via `how="left"`).

## Metro Regions

| Code | Region | Description |
|------|--------|-------------|
| `LA` | Greater Los Angeles | LA County, Orange County, Inland Empire, Ventura County |
| `SFBA` | San Francisco Bay Area | SF, Peninsula, South Bay, East Bay, North Bay, Santa Cruz |
| `GNY` | Greater New York Area | NYC boroughs, Long Island, Northern NJ, Hudson Valley |
| `Miami` | Greater Miami Area | Miami-Dade, Broward, Palm Beach counties |

## Inputs
- None (city lists are hardcoded from manual research)

## Outputs
- `data/cleaned/city_to_metro.csv`

---

## How This File Is Used

The cleaning notebooks (`02`, `03`) split each startup's `Organization Location` on comma to extract a city name, then left-join this lookup table to assign a `metro_region` value. The hazard model (`06`) uses metro region as a fixed effect.

## Parse and Combine

Convert the raw city lists into structured DataFrames and concatenate into a single lookup table.

```python
# Turn multi-line string into DataFrames
def parse_block(block, metro):
    cities = [c.strip() for c in block.splitlines() if c.strip()]
    return pd.DataFrame({"city": cities, "metro_region": metro})
```

```python
# Parse each metro region's city list into a DataFrame
df_la = parse_block(la_raw, "LA")

df_sfba = parse_block(sfba_raw, "SFBA")

df_gny = parse_block(gny_raw, "GNY")

df_miami = parse_block(miami_raw, "Miami")
```

```python
# Combined all four metro regions into a single lookup table
city_to_metro = pd.concat([df_la, df_sfba, df_gny, df_miami], ignore_index=True)
```
